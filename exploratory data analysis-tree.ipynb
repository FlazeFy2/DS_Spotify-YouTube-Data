```python
import pandas as pd
import matplotlib.pyplot as plt
import squarify

# Exploratory Data Analysis (EDA) - Tree Chart Total Song Per Context
# Dataset
data = pd.read_csv("Spotify Youtube Dataset.csv")

context_col = ['Track','Album','Channel']

for dt in context_col:
    # Count rows
    total_context = data.groupby(dt).size().reset_index(name='Count')
    
    # Sort 
    total_context = total_context.sort_values(by='Count', ascending=False).head(20)
    
    # Escape $ 
    total_context[dt] = total_context[dt].astype(str).str.replace('$', r'\$', regex=False)
    
    # Extract
    sizes = total_context['Count']
    labels = total_context.apply(lambda x: f"{x[dt]}\n({x['Count']})", axis=1)
    
    # Plot Tree Map
    plt.figure(figsize=(12, 8))
    squarify.plot(sizes=sizes, label=labels, alpha=0.8)
    plt.title(f"Total Song per {dt}", fontsize=14)
    plt.axis('off')
    plt.show()
```
