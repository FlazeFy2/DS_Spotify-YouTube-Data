```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Spotify Youtube Dataset.csv")

# Exploratory Data Analysis (EDA) - Pie Chart Most Context for Nonconditional Data
target_col = ['Album_type','Licensed','official_video']

for dt in target_col:
    # Dataset
    top_context = df[dt].value_counts()
    
    # Pie Chart
    plt.figure(figsize=(8,8))
    top_context.plot.pie(autopct='%1.1f%%', startangle=90)
    plt.title(f"Comparison Of {dt}")
    plt.ylabel("")
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
# Exploratory Data Analysis (EDA) - Pie Chart Song Distribution By Tempo Category
df = pd.read_csv("Spotify Youtube Dataset.csv")

# Convert tempo
def categorize_tempo(bpm):
    if bpm <= 60:
        return "Slow (0-60 BPM)"
    elif bpm <= 90:
        return "Ballad (61-90 BPM)"
    elif bpm <= 120:
        return "Moderate (91-120 BPM)"
    elif bpm <= 150:
        return "Fast (121-150 BPM)"
    else:
        return "Very Fast (151+ BPM)"

df["Tempo Category"] = df["Tempo"].apply(categorize_tempo)

# Count the number of missions
quartal_counts = df["Tempo Category"].value_counts().sort_index()

# Pie Chart
plt.figure(figsize=(6, 6))
quartal_counts.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title("Song Distribution By Tempo Category")
plt.ylabel("")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
# Exploratory Data Analysis (EDA) - Pie Chart Song Distribution By Duration Category
df = pd.read_csv("Spotify Youtube Dataset.csv")

df["Duration_min"] = df["Duration_ms"] / 60000

# Categorize by duration
def categorize_duration(mins):
    if mins < 2:
        return "Short (<2 min)"
    elif mins <= 4:
        return "Medium (2-4 min)"
    elif mins <= 6:
        return "Long (4-6 min)"
    else:
        return "Very Long (>6 min)"

df["Duration Category"] = df["Duration_min"].apply(categorize_duration)

# Count per category
duration_counts = df["Duration Category"].value_counts().sort_index()

# Pie chart
plt.figure(figsize=(6, 6))
duration_counts.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title("Song Distribution by Duration Category")
plt.ylabel("")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
# Exploratory Data Analysis (EDA) - Pie Chart Song Distribution By Like / Comment Ratio Category
df = pd.read_csv("Spotify Youtube Dataset.csv")

# Like Comment Ratio
df["Like_Comment_Ratio"] = df["Likes"] / (df["Comments"] + 1)

# Categorize songs by like/comment ratio
def categorize_ratio(ratio):
    if ratio < 10:
        return "Low (<10)"
    elif ratio <= 50:
        return "Medium (10-50)"
    elif ratio <= 200:
        return "High (50-200)"
    else:
        return "Viral (>200)"

df["Engagement Category"] = df["Like_Comment_Ratio"].apply(categorize_ratio)

# Count per category
engagement_counts = df["Engagement Category"].value_counts().sort_index()

# Pie chart
plt.figure(figsize=(6, 6))
engagement_counts.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title("Song Distribution by Like / Comment Ratio Category")
plt.ylabel("")
plt.tight_layout()
plt.show()
```
